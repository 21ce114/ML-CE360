```python

def temperature_category(temperature):
    if 80 <= temperature <= 90:
        return '80-90'
    elif 50 <= temperature <= 60:
        return '50-60'
    else:
        return 'Other'

df['Temperature_Category'] = df['temperature'].apply(temperature_category)


grouped_data = df.groupby('Temperature_Category')

# Display the grouped data
for category, group in grouped_data:
    print(f"Group: {category}\n")
    print(group)
    print("\n")
```
